```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
x = log_x_data
y = data['diagnosis']

label_encoder = LabelEncoder()

y_transformed = label_encoder.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(x, y_transformed, test_size = 0.2, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score, cross_val_predict

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

rfc = RandomForestClassifier(n_estimators = 100, random_state = 42)

rfc.fit(X_train_scaled, y_train)



predictions = rfc.predict(X_test_scaled)

predictions
```

```python
from sklearn.metrics import accuracy_score

test_accuracy = accuracy_score(y_test, predictions)

print(f"Test accuracy: {test_accuracy * 100:.2f}%")
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [50, 100, 200],  # Different numbers of trees
    'max_depth': [10, 20, 30],        # Different depths of trees
    'min_samples_split': [2, 5, 10]   # Different min samples for splitting
}

grid_search = GridSearchCV(estimator = rfc, param_grid = param_grid, cv = 5, scoring = 'accuracy')
```

```python
grid_search.fit(X_train, y_train)
```

```python
the_model = grid_search.best_estimator_

the_predictions = the_model.predict(X_test)
```

```python
the_test_accuracy = accuracy_score(y_test, the_predictions)

print(f"Test accuracy: {the_test_accuracy * 100:.2f}%")
```
